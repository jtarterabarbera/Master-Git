# Data retrieval


```python
from mymodule import query_tap

tap_url = "http://tap.roe.ac.uk/ssa"

adql = """
SELECT TOP 1000
    z.*,
    p.*
FROM BestDR9.ZooSpec AS z
JOIN BestDR7.PhotoObj AS p
  ON p.objid = z.dr7objid
"""

adql = """SELECT TOP 10
    z.*,
    p.*
FROM BestDR9.ZooSpec AS z
JOIN BestDR7.PhotoObj AS p
ON p.objid = z.dr7objid
AND p.objid IN (
    '587724197201248493',
    '587724197201313886',
    '587724197201379409',
    '587724197201379418',
    '587724197201379467',
    '587724197201510536',
    '587724197201510561',
    '587724197201510613',
    '587724197202296947',
    '587724197202362516'
)"""

df = query_tap(tap_url, adql).set_index("dr7objid")
```

# Image cutout


```python
from mymodule import retrieve_img
from IPython.display import Image, display

IMAGE_PIXSCALE = 0.4  # arcsec/pixel
IMAGE_SIZE_PX = 64  # in pixels
OBJECT_INDEX = 10  # which object from the filtered list to download

RA = df_filtered.iloc[OBJECT_INDEX]["ra"]
DEC = df_filtered.iloc[OBJECT_INDEX]["dec"]

blob = retrieve_img(
    ra=RA,
    dec=DEC,
    scale=IMAGE_PIXSCALE,
    width=IMAGE_SIZE_PX,
    height=IMAGE_SIZE_PX,
)

display(Image(blob))
```
